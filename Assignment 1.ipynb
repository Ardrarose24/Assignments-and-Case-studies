Bar Graph:

```python
plt.bar(data['Classification'],data['SL'])
plt.title('Classification vs SL')
plt.xlabel('Classification')
plt.ylabel('SL')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(data['Classification'],data['SW'],color = 'y')
plt.title('Classification vs SW')
plt.xlabel('Classification')
plt.ylabel('SW')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(data['Classification'],data['PL'],color = 'g')
plt.title('Classification vs PL')
plt.xlabel('Classification')
plt.ylabel('PL')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(data['Classification'],data['PW'], color = 'orange')
plt.title('Classification vs PW')
plt.xlabel('Classification')
plt.ylabel('PW')
plt.xticks(rotation=90)
plt.show()
```

Scatter plots:

```python
plt.scatter(data['PL'],data['PW'],s=30,marker = '*', color='purple')
plt.show()
```

```python
plt.scatter(data['SL'],data['SW'],s=30,marker = '*', color='purple')
plt.show()
```
