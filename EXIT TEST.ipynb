```python
# Function to plot the distribution of numerical variable
def plot_num_var(var_name, var_label, bin_num):
    plt.figure(figsize = (25,5))
    plt.suptitle('Distribution of {}'.format(var_name))
    plt.subplot(1,2,1)
    sns.histplot(data=df, x=var_name, bins=bin_num)
  #plt.title('Fig {}(a)'.format(fig_num))
    plt.xlabel(var_name)
    plt.ylabel('Count')

    plt.subplot(1,2,2)
    sns.boxplot(data=df, x=var_name)
 # plt.title('Fig {}(b)'.format(fig_num))
    plt.xlabel(var_name)
  
    plt.show()
```

```python
# Function to plot the distribution of numerical variable by selection status
def plot_numvar_selection(var_name, var_label, bin_num):
    plt.figure(figsize = (25,5))
    plt.suptitle('Distribution of {} by Selection Status'.format(var_name))
    plt.subplot(1,2,1)
    sns.histplot(data=df, x=var_name, hue='selection', bins=bin_num)
 # plt.title('Fig {}(a)'.format(fig_num))
    plt.xlabel(var_label)
    plt.ylabel('Count')
  
    plt.subplot(1,2,2)
    sns.boxplot(data=df, x='selection', y=var_name)
  #plt.title('Fig {}(b)'.format(fig_num))
    plt.xlabel('selection status')
  
    plt.show()
```

```python
for i in df.columns:                # Finding and solving outliers
    if df[i].dtype == float:
        plt.figure()
        sns.boxplot(df[i])
        plt.title(i)
        plt.show()
```

```python
num_col = ['age', 'height']
for i in num_col:
    plt.figure()
    plt.boxplot(df[i])
    plt.title(i)
```

```python
outlier_columns = ['age','height','weight','ball_controlling_skills','body_reflexes','body_balance','jumping_skills',
                   'penalties_conversion_rate','mental_strength','goalkeeping_skills','passing_skills','dribbling_skills',
                   'shot_accuracy', 'body_strength_stamina','max_running_speed','behaviour_rating','matches_played',
                   'fitness_rating','years_of_experience','no_of_disqualifications']
```

```python
for i in outlier_columns:
    
    Q1 = np.percentile(df[i], 25, interpolation = 'midpoint')
    Q2 = np.percentile(df[i], 50, interpolation = 'midpoint')
    Q3 = np.percentile(df[i], 75, interpolation = 'midpoint')
   
    
    IQR = Q3 - Q1
    
    
    up_lim = Q3+1.5*IQR
    low_lim = Q1 - 1.5*IQR
    
    
    outliers = []
    for x in df[i]:
        if(x>up_lim) or (x<low_lim):
            outliers.append(x)
            
    df[i] = np.where(df[i]>up_lim,up_lim,np.where(df[i]<low_lim,low_lim,df[i]))
```

```python
for i in outlier_columns:
    plt.figure()
    plt.boxplot(df[i])
    plt.title(i)
```

```python
#splitting data to feature and target


x = df.drop(['selection'], axis = 1)
y = df['selection']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
logit_model = LogisticRegression()
logit_model.fit(x_train, y_train)
y_pred_lr = logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('Accuracy = ',accuracy_score(y_test,y_pred_lr))
print('precision = ',precision_score(y_test,y_pred_lr))
print('Recall = ',recall_score(y_test,y_pred_lr))
print('f1 score =',f1_score(y_test,y_pred_lr))
```

```python
confusion_matrix(y_test,y_pred_lr)
```

```python
#kNN
from sklearn.neighbors import KNeighborsClassifier
metric = []
neighbors = np.arange(3,15)

for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski',p=2)
    classifier.fit(x_train,y_train)
    y_pred_knn = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_pred_knn)
    metric.append(acc)
```

```python
plt.plot(neighbors,metric,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 14, metric = 'minkowski',p=2)
classifier.fit(x_train,y_train)
y_pred_knn = classifier.predict(x_test)
```

```python
print('Accuracy = ',accuracy_score(y_test,y_pred_knn))
print('precision = ',precision_score(y_test,y_pred_knn))
print('Recall = ',recall_score(y_test,y_pred_knn))
print('f1 score =',f1_score(y_test,y_pred_knn))
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)
y_pred = dt_clf.predict(x_test)
```

```python
print('Accuracy;', accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
